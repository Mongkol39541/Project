```python
from project2 import acx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('phone.csv')
df.columns
```

```python
len(df["Phone"])
```

```python
numberphone = np.array(df["Phone"])
numberprice = np.array(df["PriceUSD"])
data_numberprice = []
data_percentphone = []
loop = 0
for num in numberphone:
    if num.isnumeric() and len(num) == 10:
        mean_num, percent_num, inpercent_num = acx.main(num)
        data_numberprice.append(df["PriceUSD"][loop])
        data_percentphone.append(percent_num)
    loop += 1
data_percentphone = np.array(data_percentphone)
data_percentphone
```

```python
len(data_percentphone)
```

```python
plt.bar(data_percentphone, data_numberprice)
plt.xlabel("Lucgy percent")
plt.ylabel("PriceUSD")
plt.show()
```
