```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#new point
x_test,y_test = make_blobs(n_samples=50,centers=4,cluster_std=0.5,random_state=0)
```

```python
x,y = make_blobs(n_samples=300,centers=4,cluster_std=0.5,random_state=0)

model = KMeans(n_clusters=4)
model.fit(x)
y_pred = model.predict(x)
y_pred_new = model.predict(x_test)
centers = model.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,1],c=y_pred)
plt.scatter(x_test[:,0],x_test[:,1],c=y_pred_new,s=120)
plt.scatter(centers[0,0],centers[0,1],c="blue",label="Centroid1")
plt.scatter(centers[1,0],centers[1,1],c="green",label="Centroid2")
plt.scatter(centers[2,0],centers[2,1],c="red",label="Centroid3")
plt.scatter(centers[3,0],centers[3,1],c="black",label="Centroid4")
plt.legend(frameon=True)
plt.show()
```
