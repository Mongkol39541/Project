```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

df = pd.read_csv("anime.csv")
DF = df[0:1000]
DF.columns
df["Licensors"][10:20]
```

```python
Licensors = df["Licensors"][0:1000]
ListLicensors1 = []
ListLicensors2 = []
ListLicensors3 = []
ListLicensors = []
Licensors_00 = 0
y = ""

for Licensors_01 in range(len(Licensors)):
    LIC = Licensors[Licensors_01]
    ListLicensors1.append(LIC)

for Licensors_02 in range(len(ListLicensors1)):
    for Licensors_03 in range(len(ListLicensors1[Licensors_02])):
        if Licensors_00 == len(ListLicensors1[Licensors_02])-1 :
            x = ListLicensors1[Licensors_02][Licensors_00]
            y += x
            ListLicensors2.append(y)
            ListLicensors3.append(y)
            break
        x = ListLicensors1[Licensors_02][Licensors_00]
        if x == ",":
            ListLicensors2.append(y)
            ListLicensors3.append(y)
            y = ""
        y += x
        if y == ",":
            y = ""
            Licensors_00 += 1
        Licensors_00 += 1
    ListLicensors.append(ListLicensors2)
    ListLicensors2 = []
    y = ""
    Licensors_00 = 0

x = set(ListLicensors3)
y = list(x)
ListLicensors3 = y
```

```python
Genres = df["Genres"][0:1000]
ListGenres1 = []
ListGenres2 = []
ListGenres3 = []
ListGenres = []
Genres_00 = 0
y = ""

for Genres_01 in range(len(Genres)):
    GEN = Genres[Genres_01]
    ListGenres1.append(GEN)

for Genres_02 in range(len(ListGenres1)):
    for Genres_03 in range(len(ListGenres1[Genres_02])):
        if Genres_00 == len(ListGenres1[Genres_02])-1 :
            x = ListGenres1[Genres_02][Genres_00]
            y += x
            ListGenres2.append(y)
            ListGenres3.append(y)
            break
        x = ListGenres1[Genres_02][Genres_00]
        if x == ",":
            ListGenres2.append(y)
            ListGenres3.append(y)
            y = ""
        y += x
        if y == ",":
            y = ""
            Genres_00 += 1
        Genres_00 += 1
    ListGenres.append(ListGenres2)
    ListGenres2 = []
    y = ""
    Genres_00 = 0

x = set(ListGenres3)
y = list(x)
ListGenres3 = y
```

```python
Name = df["Name"][0:1000]
ListName = []

Watching = df["Watching"][0:1000]
ListWatching = []

for Name_1 in range(len(Name)):
    x = Name.loc[Name_1]
    ListName.append(x)

for Watching_1 in range(len(Watching)):
    x = Watching.loc[Watching_1]
    ListWatching.append(x)
```

```python
Genres = ListGenres3 #Listของประเภท
Licensors = ListLicensors3 #Listของช่องทาง
Name = ListName #Listของชื่ออนิเมะ
Watching = ListWatching #Listของจำนวนคนดู
#Genres มีลำดับตรงกับ Check
#Name มีลำดับตรงกับ Watching
```

```python
df2 = df.sort_values("Watching",ascending=False)
NewWatching = list(df2["Watching"])
```

```python
TopWatching = NewWatching #เรียงจากมากไปน้อยของคนดู
TopName = []
TopListGenres = []
TopListLicensors = []
#ลำดับจากมากไปน้อยของชื่อ
for CheckWatching1 in range(len(TopWatching)):
    for CheckWatching2 in range(len(Watching)):
        if Watching[CheckWatching2] == TopWatching[CheckWatching1]:
            TopName.append(Name[CheckWatching2])
            TopListGenres.append(ListGenres[CheckWatching2])
            TopListLicensors.append(ListLicensors[CheckWatching2])
#Top ยังไม่ได้แยกประเภทและช่องทาง
```
